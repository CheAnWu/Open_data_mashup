### Data processing and data cleaning
The column 'price' and 'cleaning_fee', which are the price and the cleaning fee of the listing, both of them contain the dollar sign ($), so they can not be used for analysis as they are not a numerical value. I removed the dollar symbol and convert the values as numerical values by the following code.

Also, I renamed the column id to listing_id, the purpose of doing this is for creating the same column as the other datasets, and it's more convenient when merging different datasets.

Last but not least, I created a new column 'total_price' by adding the column 'price' and the column 'cleaning_fee' together.

```python
# Data Cleaning
listing_df['price'] = listing_df['price'].str.replace("[$, ]", "").astype("float")
listing_df['cleaning_fee'] = listing_df['cleaning_fee'].str.replace("[$, ]", "").astype("float")
listing_df.rename(columns={'id':'listing_id'}, inplace=True)
listing_df['total_price'] = listing_df['price'] + listing_df['cleaning_fee']
```

```python
listing_df.head()
```

I excluded some columns here, I selected 'listing_id', 'name', 'zipcode', 'latitude', 'longitude', 'property_type', 'accommodates', 'minimum_nights', 'number_of_reviews', and 'total_price'. The purpose is for mergeing with another dataframe.

```python
listing_output = listing_df[['listing_id', 'name', 'zipcode', 'latitude', 'longitude', 'property_type', 'accommodates', 'minimum_nights', 'number_of_reviews', 'total_price']]
```

### Calculate estimated revenue for each listing
If I want to calculate the estimated revenue for each listing, I should use the exact number of nights each guest stayed, but I can not get that information, so I used the listing's minimum_nights to calculate the estimated revenue. I assumed that each guest stayed at least that minimum number of nights, and for each review, I multiplied price with minimum_nights to generate the estimated revenue and each booking's revenue.

```python
pd_bookings = pd.merge(review_df, listing_output, on='listing_id')
```

Also, I summed up the revenue of every booking for each listing as estimated revenue per listing by 

"pd_listings_revenue = pd_bookings[['listing_id','estimated_revenue']].groupby(['listing_id']).sum()"

Furthermore, I merged the dataframe "listing_out" and the dataframe "pd_listing_revenue" to generate a new dataframe "pd_listings".

```python
pd_bookings['estimated_revenue'] = pd_bookings['total_price'] * pd_bookings['minimum_nights']

# get revenue by listings
pd_listings_revenue = pd_bookings[['listing_id','estimated_revenue']].groupby(['listing_id']).sum()
# pd_listings_revenue['listing_id'] = pd_listings_revenue.index

pd_listings = pd.merge(listing_output, pd_listings_revenue, on='listing_id', how='left')
pd_listings.at[pd_listings['estimated_revenue'].isnull(), 'estimated_revenue'] = 0
```

```python
from datetime import datetime
import dateutil.parser
```

### Data preprocessing
- Create an additional column "date" by using datetime.strptime function to convert the "Trip Start Timestamp" column into proper datetime format
- The reason why creating the addtional "date" column is for merging the rideshare dataset with the NBA schedule dataset

```python
for i in range(ride_df.shape[0]):
    ride_df.loc[i, "date"] = str(datetime.strptime(ride_df.loc[i, "Trip Start Timestamp"], '%m/%d/%Y %H:%M:%S %p'))
    ride_df.loc[i, "date"] = dateutil.parser.parse(ride_df.loc[i, "date"]).date()
```

```python
rideshare_nba_schedule = pd.merge(ride_df, nba, on='date', how='left')
```
